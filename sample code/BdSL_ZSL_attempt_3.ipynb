```
zero_shot_categories = ['J', 'AE', 'H', 'AB', 'E', 'V']
```

```
for d in category_vectors.keys():
    category_vectors[d] = sum(to_categorical([v-1 for v in category_vectors[d]], num_classes=20))
```

```
test_pred = model.predict(X_test)
train_pred = model.predict(X_train)
```

```
def nearest_category(vector, only_zero_shot=False):
    nearest = ""
    value = -1
    if only_zero_shot:
        categories = zero_shot_categories
    else:
        categories = category_vectors.keys() #all categories
    for k in categories:
        d = distance.euclidean(vector, category_vectors[k])
        if d < value or value == -1:
            nearest = k
            value = d
    return nearest
```

```
def hard_decision(predictions, only_zero_shot=False):
    output = []
    for pred in predictions:
        output.append(nearest_category([round(p) for p in pred], only_zero_shot))
    return output
```

```
accuracy_score(hard_decision(y_test), hard_decision(test_pred))
```

```
accuracy_score(hard_decision(y_train), hard_decision(train_pred))
```
